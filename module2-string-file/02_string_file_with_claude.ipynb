## GFF Format Recap

GFF (General Feature Format) is the standard file format for genome annotations in this lab. Key facts:

- **Tab-delimited**, 9 columns per data row
- **No header row** — comment lines start with `#` and are skipped
- **1-based** coordinate system (the first base of a chromosome is position 1)

The 9 columns in order:

| # | Name | Description |
|---|------|-------------|
| 1 | `seqname` | Chromosome / scaffold name |
| 2 | `source` | Tool or database that produced the annotation |
| 3 | `feature` | Feature type (e.g. `gene`, `exon`, `CDS`) |
| 4 | `start` | Start position (1-based, inclusive) |
| 5 | `end` | End position (1-based, inclusive) |
| 6 | `score` | Floating-point score, or `.` if absent |
| 7 | `strand` | `+`, `-`, or `.` |
| 8 | `frame` | Reading frame (0, 1, 2, or `.`) |
| 9 | `attribute` | Semicolon-separated key=value pairs |

## Loading GFF Files with pandas

Loading a GFF into a pandas DataFrame has several gotchas that trip up new users:

- **No header row** — you must pass `header=None`, otherwise pandas treats the first data line as column names.
- **Assign column names manually** after loading using `df.columns = [...]`.
- **`start` and `end` should be integers** — pandas may infer them as `int64` automatically, but verify with `df.dtypes`.
- **Skip comment lines** with `comment='#'` — this drops any `##gff-version` or `##sequence-region` header lines.

The correct loading pattern:

````python
import pandas as pd

columns = ['seqname', 'source', 'feature', 'start', 'end',
           'score', 'strand', 'frame', 'attribute']

df = pd.read_csv('your_file.gff', sep='\t', header=None, names=columns, comment='#')
df.head()
````
